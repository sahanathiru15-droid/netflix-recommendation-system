```python
movies=movies[['show_id','title','description','rating','listed_in']]
movies
```

```python
movies['tags']=movies['description']+movies['listed_in'].copy()
movies
```

```python
new_data=movies.drop(columns=['description','listed_in'])
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=7787,stop_words='english')
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
similarity
```

```python
new_data[new_data['title']=="23:59"].index[0]
```

```python
new_data.info()
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
 print(new_data.iloc[i[0]].title)
```

```python
def recommand(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommand("A Serious Man")
```
